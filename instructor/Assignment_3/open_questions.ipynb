Lastly,weighted loss pushes model to focus on minority classes, this can make neutral predictions noisier.

# 3. Cross-Model Comparison

    1.How did mean-pooled FastText limit the MLP?
MLP uses mean-pooled FastText embeddings, so the mean-pooled FastText will averages all word vectors, which loses word order and context. This means that the mean-pooled method can not capture the subtle tone.Thus, this performs worse than sequence models that read sentences word by word.

    2. What advantage did LSTM have over MLP?
LSTM will process the text sequentially, so it will keep information in word order and word dependencies. It can also learn patterns like â€œnot goodâ€ or contrast phrases. This will help LSTM to model sentence meaning better than the simple averaging.

    3. Did fine-tuned LLMs outperform classical models? Why?
Yes, BERT and GPT clearly performed best. This is because they were pretrained on massive scale of text dataset, so they understand language patterns before the fine-tuned. Their contextual embeddings capture word meaning which depends on surrounding words. This can improve the overall sentiment classification.
